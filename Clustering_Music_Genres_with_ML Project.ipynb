#Project Statement

Music preferences are hard to predict from personal traits like lifestyle or hobbies. However, by analyzing the audio features of songs a user listens to, we can identify patterns and recommend similar music. This project focuses on clustering Spotify songs based on their audio characteristics to group similar music genres.


```python
# Import Libaries
import pandas as pd
import numpy as np
from sklearn import cluster

# Loading dataset
data = pd.read_csv("/content/Spotify-2000.csv")
data.head()
```

```python
#Drop the Index
data = data.drop("Index", axis=1)
data
```

```python
# Correlation between Numeric Values
data.corr(numeric_only=True)
```

```python
#K Means Clustering (Similarities between all the audio features)
data2 = data[["Beats Per Minute (BPM)", "Loudness (dB)",
              "Liveness", "Valence", "Acousticness",
              "Speechiness"]]
data2
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
for i in data.columns:
    MinMaxScaler(i)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
clusters = kmeans.fit_predict(data2)
```

```python
data["Music Segments"] = clusters
MinMaxScaler(data["Music Segments"])
data["Music Segments"] = data["Music Segments"].map({1: "Cluster 1", 2:
    "Cluster 2", 3: "Cluster 3", 4: "Cluster 4", 5: "Cluster 5",
    6: "Cluster 6", 7: "Cluster 7", 8: "Cluster 8",
    9: "Cluster 9", 10: "Cluster 10"})
```

```python
import plotly.express as px

fig = px.scatter_3d(
    data,
    x='Beats Per Minute (BPM)',
    y='Energy',
    z='Danceability',
    color='Music Segments'
)

fig.show()
```
