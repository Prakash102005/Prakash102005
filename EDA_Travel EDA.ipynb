Key Insights from the Statistical Summary:

Target Variable (ProdTaken) Distribution:

About 29.2% of customers purchased the product (ProdTaken = 1).
The dataset is imbalanced, with 70.8% of customers not purchasing.

Customer Age:

Ranges from 18 to 61 years, with a mean of 37.6 years.
226 missing values need imputation.

Sales Pitch & Engagement:

Duration of Pitch: Avg 15.5 minutes, but ranges from 5 to 127 minutes.
Number of Follow-ups: Most have 3-4 follow-ups, but some have up to 6.
Pitch Satisfaction Score: Mostly around 3 (neutral to satisfied range).

Travel Preferences:

Preferred Property Star Rating: Most prefer 3-4 stars.
Number of Trips: Customers take about 3 trips on average, but some have taken up to 22 trips.

Financial & Lifestyle Factors:

Monthly Income: Varies widely, with an average of ₹23,620.
Own Car: 63.5% of customers own a car.
Passport: Only 37% have a passport, indicating a potential for international travel targeting.

Missing Data:

DurationOfPitch, MonthlyIncome, Age, and NumberOfTrips have notable missing values.
Type of Contact (25 missing) and PreferredPropertyStar (26 missing) also need attention.

## Method 2: Label Encoding

In label encoding in Python, we replace the categorical value with a numeric value between 0 and the number of classes minus 1. If the categorical variable value contains 5 distinct classes, we use (0, 1, 2, 3, and 4)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Occupation_Encoded'] = le.fit_transform(cat_data['Occupation'])
df.head()
```

```python
# Print the mapping of classes to numerical values
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(label_mapping)
```
