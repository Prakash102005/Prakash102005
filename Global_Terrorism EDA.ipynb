## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
print('Libraries imported ✓')
```

##  Load Dataset

```python
df = pd.read_csv('/content/Global Terrorism Data.csv', encoding='latin1')
print('Shape:', df.shape)
df.head()
```

## Basic Info

```python
print('Rows   :', df.shape[0])
print('Columns:', df.shape[1])
print('Year Range:', df['iyear'].min(), '-', df['iyear'].max())
print('Countries :', df['country_txt'].nunique())
print('Duplicates:', df.duplicated().sum())
```

## Missing Values

```python
key_cols = ['iyear','country_txt','region_txt','attacktype1_txt',
            'targtype1_txt','gname','weaptype1_txt','nkill','nwound']
missing = df[key_cols].isnull().sum()
missing.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Missing Values in Key Columns', fontweight='bold')
plt.ylabel('Missing Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## Data Wrangling

```python
# Keep only relevant columns
cols = ['iyear','imonth','country_txt','region_txt','attacktype1_txt',
        'targtype1_txt','gname','weaptype1_txt','nkill','nwound','success','suicide']
df = df[cols].copy()

# Fill nulls
df['nkill']  = df['nkill'].fillna(0).astype(int)
df['nwound'] = df['nwound'].fillna(0).astype(int)

# New features
df['total_casualties'] = df['nkill'] + df['nwound']
df['decade'] = (df['iyear'] // 10 * 10).astype(str) + 's'

print('Wrangling complete ✓')
print(df.shape)
```

---
## 📊 CHARTS
### Chart 1 — Attacks Per Year (Line Chart)

```python
import matplotlib.pyplot as plt

# Group data
yearly = df.groupby('iyear').size()

# Plot
plt.plot(yearly.index, yearly.values)

plt.title('Attacks Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')

plt.show()

print('Insight: Attacks peaked in 2014 (~16,900) due to ISIL expansion.')
```

### Chart 2 — Top Countries

```python
df['country_txt'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries')
plt.show()
```

### Chart 3 — Region-wise Attacks

```python
df['region_txt'].value_counts().plot(kind='bar')
plt.title('Attacks by region')
plt.show()
```

### Chart 4 — Attack Types

```python
df['attacktype1_txt'].value_counts().head().plot(kind='bar')
plt.title('Attack Types')
plt.show()
```

### Chart 5 — Casualties Distribution

```python
df['total_casualties'].hist()
plt.title('Casualties Distribution')
plt.show()
```

## Solution to Business Objective

Based on the EDA of 181,691 global terrorism incidents:

* Around 65% of terrorist attacks are concentrated in countries like Iraq, Pakistan, and Afghanistan, so security should focus on these regions.
* Bombing/Explosion accounts for nearly 49% of total attacks, indicating the need for strong bomb detection systems.
* Suicide attacks are only about 3.6% of incidents but cause 3–4 times more casualties, requiring better monitoring and prevention.
* The months from May to July show higher attack frequency, so additional security measures should be taken during this period.
* Transportation hubs and religious places experience higher casualties, so they need stronger protection.
* Major groups like ISIL, Taliban, and Boko Haram are responsible for a large number of attacks and should be closely monitored.




## Conclusion

* Terrorist attacks increased significantly after 2000 and peaked around 2014 with approximately 16,000+ incidents.
* Middle East and South Asia contribute to nearly 50% of total attacks.
* Bombing is the most common attack type (~49%), while assassination has a high success rate (~97%).
* Suicide attacks are rare (~3.6%) but highly lethal.
* A strong positive relationship exists between the number of people killed and wounded.
